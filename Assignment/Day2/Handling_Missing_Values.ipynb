Ques 1) How to handle missing data. Reading a data from notepad file

```python
import pandas as pd
autos = pd.read_csv("/content/auto-mpg.data",
                    sep='\s+',
                    header=None)
autos.head(5)
```

```python
autos.columns = ['mgp','cylinders','displacement','horespower','weight','acceleration','year','origin','name']
autos.head(5)
```

```python
autos.info()
```

```python
autos["horespower"] = pd.to_numeric(autos["horespower"], errors ='coerce')
autos.info()
```

```python
autos[autos.horespower.isnull()]
```

```python
autos = autos.dropna(subset = ["horespower"])
autos[autos.horespower.isnull()]
```
