Ques 1) Load Stock1 into dataframe. Display their 5 entries

```python
import pandas as pd
df = pd.read_csv("/content/BEML.csv")
dfw = pd.read_csv("/content/GLAXO.csv")
print(df.head(5))
print(dfw.head(5))
```

Ques 2) Select a few attributes which is needed to check performance of two attribute

```python
df1 = df[["Date","Close"]]
df2 = dfw[["Date","Close"]]
print(df1.head(5))
print(df2.head(5))
```

Ques 3) Preparing a time plot for a close attribute

```python
import matplotlib.pyplot as plt
import seaborn as sn
%matplotlib inline

plt.plot(df1.Close);
plt.plot(df2.Close);
plt.xlabel("Time");
plt.ylabel("Close Price")
```

Ques 4) Arrange the data in the order of time

```python
df1 = df1.set_index(pd.DatetimeIndex(df1["Date"]))
print(df1.head(5))
df2 = df2.set_index(pd.DatetimeIndex(df2["Date"]))
df2.head(5)
```

```python
plt.plot(df1.Close);
plt.plot(df2.Close);
plt.xlabel("Time");
plt.ylabel("Close Price")
```

Ques 5) The dataframe which is created df1 move into newbeml csv file

```python
destination_file = "NewBEML.csv"
df1.to_csv(destination_file, index=False)
print("ETL process completed")
destination_file1 = "NewGLAXO.csv"
df2.to_csv(destination_file1, index=False)
print("ETL1 process completed")
```

```python
df1['gain'] = df1.Close.pct_change(periods =1)
df2['gain'] = df2.Close.pct_change(periods =1)
print(df1.head(5))
df2.head(5)
```

```python
plt.figure(figsize = (8,6))
plt.plot(df1.index,df1.gain)
plt.plot(df2.index,df2.gain)
plt.xlabel("Time")
plt.ylabel("Gain")
```

```python
sn.distplot(df2.gain, label = "Glaxo")
sn.distplot(df1.gain, label = 'BEML')
plt.xlabel('gain')
plt.ylabel('Density')
plt.legend()
```

```python
print("Daily gain of Glaxo")
print("-------------------------")
print("mean: ",round(df2.gain.mean(),4))
print("Standard Deviation ",round(df2.gain.std(),4))
print("***********************************")
print("Daily gain of BEML")
print("-------------------------")
print("mean: ",round(df1.gain.mean(),4))
print("Standard Deviation ",round(df1.gain.std(),4))
```

```python
print(df1.gain.describe())
df2.gain.describe()
```

```python
import numpy as np
from scipy import stats
glaxo_df_ci = stats.norm.interval(0.95,
                                  loc = df2.gain.mean(),
                                  scale = df2.gain.std())
print("Gain at 95% confidence interval in GLAXO is : ",np.round(glaxo_df_ci,4))
beml_df_ci = stats.norm.interval(0.95,
                                  loc = df1.gain.mean(),
                                  scale = df1.gain.std())
print("Gain at 95% confidence interval in BEML is : ",np.round(beml_df_ci,4))
```
