Ques 1) A passport office claims that the passport applications are processed within 30days of submitting the application form and all
necessary documents. The file passport.csv contains processing time of 40 passport applicants. The population standard deviation of the processing time is 12.5 days. Conduct a hypothesis test at significance level ∝ =0.05 to verify the claim made by the passport office.

```python
import pandas as pd
passport_df = pd.read_csv("passport.csv")
passport_df.head(5)
```

```python
import math
def z_test(pop_mean, pop_std, sample):
  z_score = (sample.mean() - pop_mean) / (pop_std/math.sqrt(len(sample)))
  return z_score, stats.norm.cdf(z_score)
```

```python
z_test(30,12.5,passport_df.processing_time)
```

Ques 2) Aravind Productions (AP) is a newly formed movie production house based out of Mumbai, India. AP was interested in understanding the production cost required for producing Bollywood movie. The industry believes that the production house will require INR 500 million on average. It is assumed that the Bollywood movie production cost follows a normal distribution. The production costs of 40 Bollywood movies in millions of rupees are given in bollywoodmovies.csv file. Conduct and appropriate hypothesis test at ∝ =0.05 to check whether the belief about average production cost is correct.

```python
bollywood_movies_df = pd.read_csv('bollywoodmovies.csv')
bollywood_movies_df.head(5)
```

```python
from scipy import stats
stats.ttest_1samp(bollywood_movies_df.production_cost,500)
```

Ttest_1sampResult(statistics = -2.2845, pvalue = 0.02786)

Ques 4) A company claims that children who drink their health drink will grow taller than the children who do not drink that health drink. Data in the file healthdrink.xlsx shows average increase in height over one-year period from two groups: one drinking the health drink and the other not drinking the health drink. At ∝ =0.05, test whether the increase in height for the children who drink the health drink is different than those who do not drink health drink.

```python
healthdrink_yes_df = pd.read_excel("healthdrink.xlsx",'healthdrink_yes')
healthdrink_yes_df.head(5)
```

```python
healthdrink_no_df = pd.read_excel("healthdrink.xlsx",'healthdrink_no')
healthdrink_no_df.head(5)
```

```python
stats.ttest_ind(healthdrink_yes_df['height_increase'],healthdrink_no_df['height_increase'])
```

Ttest_indResult(statistics = 8.1316, pvalue = 0.00)

Ques 5) The file breakup.csv contains alcohol consumption before and after breakup. Conduct a paired t-test to check whether the alcohol consumption is same after the breakup at 95% confidence(∝ =0.05).

```python
breakups_df = pd.read_csv('breakups.csv')
breakups_df.head(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
sn.distplot(breakups_df['Before_Breakup'],label='Before_Breakup')
sn.distplot(breakups_df['After_Breakup'],label='After_Breakup')
plt.legend()
```

```python
stats.ttest_rel(breakups_df['Before_Breakup'],breakups_df['After_Breakup'])
```

Ttest_relResult(statistic = -0.5375, pvalue= 0.5971)
